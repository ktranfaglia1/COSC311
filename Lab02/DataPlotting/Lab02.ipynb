### Task 1
Use pandas package to:
- use info() function to show the column information of these two datasets separately
- use describe() function to show the statistics (count, mean, std, min, ...) of these two datasets separately
- show the unique values of the "Ws" (wind speed) column of these two datasets separately
- count how many samples there are in each of these two datasets separately

```python
import pandas as pd

# Read in data 
Bejaia_Region = pd.read_csv('Bejaia_Region.csv')
SidiBel_Abbes_Region =pd.read_csv('Sidi-Bel_Abbes_Region.csv')

# Display info for each dataset: shows the column information
print(Bejaia_Region.info(), "\n")
print(SidiBel_Abbes_Region.info(), "\n")

# Display descriptive statistics for each dataset: Count, mean, std, min/max, quartiles
print(Bejaia_Region.describe(), "\n")
print(SidiBel_Abbes_Region.describe(), "\n")

# Display unique values for 'Ws' in each dataset: 
print(Bejaia_Region['Ws'].unique(), "\n")
print(SidiBel_Abbes_Region['Ws'].unique(), "\n")

# Display the count for each dataset: Number of samples for each column
print(Bejaia_Region.count(), "\n")
print(SidiBel_Abbes_Region.count(), "\n")
```

### Task 2
Draw a line figure to show the temperature change with time for the "Bejaia Region Dataset"


```python
import matplotlib.pyplot as plt

# Get a Series for temperature by extracting temperature data and a Time Stamp Series for time by using pandas to combine day, month, and year once extracted
temperature = Bejaia_Region['Temperature'] 
time = pd.to_datetime(Bejaia_Region[['day', 'month', 'year']])

# Setup and plot data for a line graph
plt.figure(figsize=(10, 6))  # Set figure size
plt.plot(time, temperature, marker='o', linestyle='-')  # Plot Dotted Line Graph with markers at data points
plt.title('Temperature Change Over Time')
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.grid(True)  # Set plot to be a grid for easier identification
plt.tight_layout()  # Adjust layout to prevent clipping of labels, specifically for x labels
plt.show()
```

### Task 3
Draw a scatterplot figure to show the relationship between the temperature and the Fire Weather Index (FWI) for the "Sidi-Bel Abbes Region Dataset"


```python
# Get a Series for temperature and a Series for fwi by extracting the data
temperature = SidiBel_Abbes_Region['Temperature']
fwi = SidiBel_Abbes_Region['FWI']

# Setup and plot data for a scatterplot
plt.figure(figsize=(10, 6))  # Set figure size
plt.scatter(temperature, fwi, color='purple')  # Plot a scatterplot
plt.title('Relationship Between Temperature and Fire Weather Index (FWI)')
plt.xlabel('Temperature')
plt.ylabel('FWI')
plt.grid(True)  # Set plot to be a grid for easier identification
plt.show()
```

### Task 4
Draw a histogram to show the average Relative Humidity (RH) for each month for the "Bejaia Region Dataset"

```python
months = ['Jun', 'Jul', 'Aug', 'Sep']  # List of months in dataset
avg_rh = Bejaia_Region.groupby('month')['RH'].mean()  # Calculate average rh by month

# Setup and plot data for a bar chart
plt.figure(figsize=(10, 6))  # Set figure size
plt.bar(months, avg_rh, color='teal', edgecolor='black')  # Assign bar chart to represent the data
plt.title('Average Relative Humidity (RH) by Month')
plt.xlabel('Month')
plt.ylabel('Average RH')
plt.grid(True)  # Set plot to be a grid for easier identification
plt.show()
```

Draw a bar figure to show the maximum Rain amount in a day for each month for the "Bejaia Region Dataset"


```python
months = ['Jun', 'Jul', 'Aug', 'Sep']  # List of months in dataset
max_rain = Bejaia_Region.groupby('month')['Rain'].max()  # Calculate max daily rain by month

# Setup and plot data for a bar chart
plt.figure(figsize=(10, 6))  # Set figure size
plt.bar(months, max_rain, color='violet', edgecolor='black')  # Assign bar chart
plt.title('Maximum Rain Amount in a Day by Month')
plt.xlabel('Month')
plt.ylabel('Maximum Rain')
plt.grid(True)  # Set plot to be a grid for easier identification
plt.show()
```

### Task 6
Draw a histogram to show the Wind speed (Ws) distribution in 5 bins for the "Sidi-Bel Abbes Region Dataset" in June, 2012

```python
# Setup and plot data for a histogram
plt.figure(figsize=(10, 6))  # Set figure size
plt.hist(SidiBel_Abbes_Region['Ws'][SidiBel_Abbes_Region['month'] == 6], bins=5)  # Assign data for a histogram
plt.xlabel('Wind Speed')
plt.ylabel('Frequency (# of days)')
plt.title("Wind Speed (Ws) Distribution in June, 2012")
plt.grid(True)  # Set plot to be a grid for easier identification
plt.show()
```

### Task 7
Draw a line figure to show the correlation between temperature (Temp) and Relative Humidity (RH) for the "Sidi-Bel Abbes Region Dataset" in July, 2012

```python
import numpy as np

days_in_july = np.arange(1, 32)  # Days for entire month of July

# Set up and plot line graph
plt.figure(figsize=(10, 6))  # Set figure size
plt.plot(days_in_july, SidiBel_Abbes_Region['Temperature'][SidiBel_Abbes_Region['month'] == 7], 'r-', label='Temperature')  # red solid line for Temp in July, 2012
plt.plot(days_in_july, SidiBel_Abbes_Region['RH'][SidiBel_Abbes_Region['month'] == 7], 'g-.', label='RH')  # green dot-dashed line for RH in July, 2012
plt.legend(loc=9)  # Create legend and place it at the top of the figure
plt.xlabel("Days")
plt.ylabel(" Temperature | RH")
plt.title("Correlation Between Temperature and Relative Humidity (RH) in July, 2012")
plt.grid(True)  # Set plot to be a grid for easier identification
plt.show()
```

### Task 8
Draw a bar figure to show the distribution of Relative Humidity (RH) for the "Bejaia Region Dataset". The x-axis is the decile of Relative Humidity
(20s, 30s, ..., 90s), and y-axis is the number of days

```python
from collections import Counter

# Bucket RH by decile, but put 100 in with the 90s (Does not have any impact on this data set)
rh_decile = Counter(min(x // 10 * 10, 90) for x in Bejaia_Region['RH'])

# Setup and plot data for a bar chart
plt.figure(figsize=(10, 6))  # Set figure size
plt.bar([x + 5 for x in rh_decile.keys()], rh_decile.values(), 10, color='violet', edgecolor='black')  # Assign bar chart: Shift bars right 5, give height data, set width 10
plt.xticks([10 * i for i in range(11)])  # Set x-axis labels at 0-100 incrementing by 10
plt.title('Relative Humidity Distribution')
plt.xlabel('Relative Humidity')
plt.ylabel('Number of Days')
plt.grid(True)  # Set plot to be a grid for easier identification
plt.show()
```

### Task 9
Draw a figure (any type you want) to show the average temperature for each month when there is "no fire" and there is "fire" for the "Bejaia Region Dataset"

```python
months = ['Jun', 'Jul', 'Aug', 'Sep']  # List of months in dataset

# Calculate average temperature for each month when there is no fire and when there is fire
no_fire_avg_temp = Bejaia_Region[Bejaia_Region['Classes'] == 'not fire   '].groupby('month')['Temperature'].mean()
fire_avg_temp = Bejaia_Region[Bejaia_Region['Classes'] == 'fire   '].groupby('month')['Temperature'].mean()

# Set up and plot data for a line graph
plt.figure(figsize=(10, 6))  # Set figure size
plt.plot(months, no_fire_avg_temp, marker='o', color='blue', label='No Fire')  # Plotting the average temperature for "not fire" with solid blue line 
plt.plot(months, fire_avg_temp, marker='o', color='red', label='Fire')  # Plotting the average temperature for "fire" with solid red line
plt.xlabel('Month')
plt.ylabel('Average Temperature')
plt.title('Average Temperature for Each Month with and without Fire')
plt.legend()  # Create a legend for the line graph
plt.grid(True)  # Set plot to be a grid for easier identification
plt.show()

# Note: I wanted to try it with a bar graph too, but this one required some research. I wanted the bars to be spaced out for comparison instead of overlayed
x = np.arange(len(months))  # Get evenly spaced values for x-axis positions using number (length) of months
bar_width = 0.4  # The width for each bar

# Set up and plot data for a bar chart
plt.figure(figsize=(10, 6))  # Set figure size
plt.bar(x - bar_width/2, no_fire_avg_temp, bar_width, color='blue', label='No Fire')  # Plotting bars for "not fire" average temperatures with blue bar
plt.bar(x + bar_width/2, fire_avg_temp, bar_width, color='red', label='Fire')  # Plotting bars for "fire" average temperatures with red bar
plt.xlabel('Month')
plt.ylabel('Average Temperature')
plt.title('Average Temperature for Each Month with and without Fire')
plt.xticks(x, months)  # Set x-axis ticks to be at positions x (evenly spaced values) with months as labels
plt.legend()  # Create a legend for the bar chart
plt.grid(True)  # Set plot to be a grid for easier identification
plt.show()
```
